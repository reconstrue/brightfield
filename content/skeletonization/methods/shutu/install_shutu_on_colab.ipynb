# Install ShuTu on Colab

The goal of this notebook is simply to install ShuTu SWC generator on Colab. The ShuTu SWC editor desktop app is out of scope.

This install script is based on [the ShuTu tutorial](http://personal.psu.edu/dzj2/ShuTu/) but things are tweaked and tuned for installation on Colab. (The default instructions raise a harmless but confusing error on Colab.) See [test_install_of_shutu_on_colab.ipynb](https://colab.research.google.com/drive/1PQEWuFjUi3vo-1btNi6M5vXoNS3dTv5o#scrollTo=4xD6MAvUX1lf), for a blow-by-blow log of how this script was developed for deploying ShuTu, but specifically on Colab.

ShuTu only needs to be installed once per VM run (max of 12 hours) so it is useful to have installation be a separate notebook.


ShuTu can be installed using the default method as run through in [brightfield_neuron_swc_by_shutu_on_colab.ipynb](https://colab.research.google.com/drive/1PQEWuFjUi3vo-1btNi6M5vXoNS3dTv5o#scrollTo=qQ5T40Rh0gJk). It works but generates a harmless error at the end of build.sh. So instead, here the core of `build.sh` is recreated as a Colab code cell. For an explanation see [GitHub Issues](https://github.com/reconstrue/brightfield_neuron_reconstruction/issues/5#issuecomment-541471946). Basically, the build.sh boils down to calling make twice in the appropriate directories.
